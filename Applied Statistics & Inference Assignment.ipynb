## Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

### Ans: Type I Error (α Error)-A Type I error happens when we reject a true null hypothesis — i.e., we conclude an effect exists when it does not.

* Symbol: α (alpha)

* Example: Approving a drug that actually has no effect.

* Impact: Leads to false positives or wrong claims.

* Control: The probability of this error equals the significance level (α), commonly set at 0.05.

### Type II Error (β Error)-A Type II error occurs when we fail to reject a false null hypothesis — i.e., we miss a real effect.

* Symbol: β (beta)

* Example: Rejecting a truly effective drug.

* Impact: Leads to false negatives, missing genuine findings.

* Power of the test: (1 − β) shows the ability to detect a true effect.

### **Impact on Decision-Making**

* Type I Error: May cause acceptance of false claims — costly in medicine, research, and policy.

* Type II Error: May overlook real discoveries or useful interventions.Hence, researchers must balance both errors based on the situation’s seriousness.

## Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

### Ans:The **P-value** is the **probability of obtaining the observed results (or more extreme results)** assuming that the **null hypothesis is true**.It quantifies how compatible the sample data are with H₀.

* **Formulaically:**
  ( P = P(\text{data or more extreme data} \mid H₀ \text{ is true}) )

### **Interpretation**

* A **small P-value** (typically ≤ 0.05) indicates that such data would be very unlikely if H₀ were true → **strong evidence against H₀**, so we **reject H₀**.
* A **large P-value** (> 0.05) suggests that the data are consistent with H₀ → **insufficient evidence to reject H₀**.

| **P-value** | **Interpretation**              | **Decision**      |
| ----------- | ------------------------------- | ----------------- |
| ≤ 0.01      | Very strong evidence against H₀ | Reject H₀         |
| ≤ 0.05      | Moderate evidence against H₀    | Reject H₀         |
| > 0.05      | Weak or no evidence against H₀  | Fail to reject H₀ |

### **Example**

Suppose we test whether a new medicine lowers blood pressure.
If **P = 0.03**, it means there’s a 3% chance of observing such data if the drug truly has **no effect**. Since 0.03 < 0.05, we reject H₀ and conclude the medicine likely works.

### **Common Misinterpretations**

* A **P-value does not** show the probability that H₀ is true or false.
* It only indicates how unusual the observed data are **under the assumption that H₀ is true**.
* A low P-value ≠ proof of effect; a high P-value ≠ proof of no effect.

### **Conclusion**

The P-value is a **measure of statistical evidence** against the null hypothesis. A smaller P-value implies stronger evidence to reject H₀, while a larger P-value suggests insufficient evidence. Proper interpretation ensures sound and objective decision-making in research.

## Question 3:Explain the difference between a Z-test and a T-test, including when to use each.

### Ans:Both **Z-test** and **T-test** are statistical methods used to determine whether there is a significant difference between sample data and a population or between two samples. The main difference lies in the **sample size** and whether the **population standard deviation (σ)** is known.

### **1. Z-Test**

A **Z-test** is used when:

* The **population standard deviation (σ)** is **known**.
* The **sample size (n)** is **large** (usually n > 30).
* The data are **normally distributed**.

**Formula:**
[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
]
Where:

* (\bar{X}) = sample mean
* (\mu) = population mean
* (\sigma) = population standard deviation
* (n) = sample size

**Example:**
Testing whether the average height of 100 students differs from the national average when population variance is known.

### **2. T-Test**

A **T-test** is used when:

* The **population standard deviation (σ)** is **unknown**.
* The **sample size** is **small** (n < 30).
* Data are approximately **normally distributed**.

**Formula:**
[
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
]
Where:

* (s) = sample standard deviation (used instead of σ)

**Example:**
Testing whether the average marks of 20 students differ from the expected mean when σ is unknown.

**Key Differences**

| **Basis**                     | **Z-Test**                                     | **T-Test**                                     |
| ----------------------------- | ---------------------------------------------- | ---------------------------------------------- |
| Population standard deviation | Known                                          | Unknown                                        |
| Sample size                   | Large (n > 30)                                 | Small (n < 30)                                 |
| Distribution used             | Normal (Z-distribution)                        | Student’s t-distribution                       |
| Variability                   | Less (σ fixed)                                 | More (depends on sample)                       |
| Example                       | Comparing large sample mean to population mean | Comparing small sample mean to population mean |


### Note:

* Use **Z-test** for **large samples** or when σ is known.
* Use **T-test** for **small samples** or when σ is unknown.
* As sample size increases, the **T-distribution** approaches the **Z-distribution**.

### **Conclusion**

Both tests serve the same purpose—testing hypotheses about means—but differ in assumptions. The **Z-test** is ideal for large samples with known population variance, while the **T-test** is suited for small samples where the variance must be estimated. Correct test selection ensures valid and reliable statistical conclusions.

## Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?

### Ans:A **confidence interval (CI)** is a range of values, derived from sample data, that is likely to contain the **true population parameter** (such as mean or proportion) with a specified level of confidence. It provides both an estimate and a measure of uncertainty in statistical inference.A **confidence interval** gives an estimated range within which the true population value lies, based on a chosen **confidence level** (commonly 90%, 95%, or 99%).

**Formula:**
[
\text{Confidence Interval} = \bar{X} \pm Z \left(\frac{\sigma}{\sqrt{n}}\right)
]
or when σ is unknown,
[
\bar{X} \pm t \left(\frac{s}{\sqrt{n}}\right)
]

Where:

* (\bar{X}) = sample mean
* (Z) or (t) = critical value
* (\sigma) or (s) = standard deviation
* (n) = sample size

### **Interpretation**

A **95% confidence interval** means that if we were to take 100 similar samples, about **95 of them** would contain the true population mean.
It does **not** mean there’s a 95% probability that the mean lies within a single interval — the parameter is fixed, the interval varies.

### **Margin of Error (E)**

The **margin of error** represents the maximum expected difference between the sample estimate and the true population value.

[
E = Z \left(\frac{\sigma}{\sqrt{n}}\right)
]

It directly determines the **width** of the confidence interval:
[
\text{Width of CI} = 2E
]


### **Effect on Width and Interpretation**

* **Larger Margin of Error → Wider CI:**
  Less precise but more confident (e.g., 99% CI).
* **Smaller Margin of Error → Narrower CI:**
  More precise but less confident (e.g., 90% CI).
* **Factors affecting margin of error:**

  * **Confidence level:** Higher confidence → larger margin.
  * **Sample size:** Larger sample → smaller margin.
  * **Data variability:** Greater variability → larger margin.

### **Example**

If the mean score is 70 with a 95% CI of (67, 73), we are 95% confident that the true mean lies between **67 and 73**.
A smaller margin of error (say ±1) would narrow the range, increasing precision.

### **Conclusion**

A **confidence interval** provides both an estimate and a reliability measure for population parameters. The **margin of error** controls the interval’s width — balancing **precision** and **confidence**. Wider intervals give more confidence but less precision, while narrower intervals increase precision but reduce confidence.

## Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

### Ans:**ANOVA (Analysis of Variance)** is a statistical test used to determine whether there are **significant differences between the means** of **three or more groups**. It extends the concept of the **t-test**, which is limited to comparing only two groups, by analyzing overall variance in the data.

### **Purpose of ANOVA**

The main purpose of ANOVA is to:

* Test whether **group means differ significantly**.
* Separate total variability into **between-group variance** and **within-group variance**.
* Identify if any observed differences among group means are **statistically significant** or just due to **random chance**.

**Hypotheses in ANOVA:**

* **Null Hypothesis (H₀):** All group means are equal.
  [
  H₀: \mu_1 = \mu_2 = \mu_3 = \dots = \mu_k
  ]
* **Alternative Hypothesis (H₁):** At least one group mean differs.

---

### **How ANOVA Works**

ANOVA compares the **ratio of variances**:
[
F = \frac{\text{Between-group variance}}{\text{Within-group variance}}
]

* If the **F-value** is large (and **P-value < 0.05**), we reject H₀, meaning not all means are equal.
* If **F-value** is small, we fail to reject H₀.

### **Assumptions of ANOVA**

To ensure valid results, ANOVA relies on the following assumptions:

1. **Independence:** Observations are independent of each other.
2. **Normality:** Data in each group are approximately normally distributed.
3. **Homogeneity of variances:** All groups have equal variances.
   (Tested using Levene’s or Bartlett’s test.)

Violating these assumptions can lead to incorrect conclusions.

### **Extension Beyond Two Groups**

* A **t-test** compares only **two means**, increasing Type I error if repeated.
* **ANOVA** allows testing **multiple groups simultaneously** (e.g., comparing 3 or more treatment means) while controlling overall error rate.
* If ANOVA shows significant differences, **post-hoc tests** (like Tukey’s or Bonferroni) identify **which specific groups differ**.

### **Example**

A researcher compares the average exam scores of students taught with **three different teaching methods**.
ANOVA determines if at least one method leads to significantly different average performance.

### **Conclusion**

ANOVA is a powerful extension of hypothesis testing that compares **more than two means** efficiently. By analyzing variance and maintaining statistical control, it helps researchers identify meaningful group differences while minimizing the risk of error.

## Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

```python
# One-Sample Z-Test in Python

import numpy as np
from scipy import stats

# Sample data (e.g., test scores of 50 students)
data = [65, 70, 68, 75, 72, 66, 71, 69, 74, 73,
        67, 70, 69, 72, 68, 65, 71, 70, 73, 74,
        66, 68, 70, 72, 69, 75, 67, 74, 71, 73,
        68, 69, 70, 72, 71, 73, 74, 75, 68, 69,
        71, 72, 70, 74, 73, 69, 70, 72, 71, 73]

# Given population parameters
population_mean = 70     # μ₀
population_std = 3       # σ

# Calculate sample statistics
sample_mean = np.mean(data)
n = len(data)

# Compute Z statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Compute two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.3f}")
print(f"P-value = {p_value:.4f}")

# Interpretation at α = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H₀: The sample mean is significantly different from the population mean.")
else:
    print("Fail to Reject H₀: No significant difference between sample and population mean.")
```

## Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate non-normal population (Exponential distribution)
population = np.random.exponential(scale=2.0, size=100000)

# Step 2: Draw multiple samples and compute their means
sample_means = []
sample_size = 50  # size of each sample
num_samples = 1000  # number of samples

for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

# Step 3: Plot histograms
plt.figure(figsize=(12,5))

# Original non-normal population
plt.subplot(1,2,1)
plt.hist(population, bins=40, color='skyblue', edgecolor='black')
plt.title("Original Non-Normal Distribution (Exponential)")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Sampling distribution of means
plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title("Sampling Distribution of Means (Approx. Normal)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval(data, confidence=0.95):
    # Step 1: Calculate sample statistics
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error = s / sqrt(n)
    
    # Step 2: Compute t-critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    
    # Step 3: Compute margin of error and CI
    margin = t_crit * std_err
    lower = mean - margin
    upper = mean + margin
    
    # Print results
    print(f"Sample Mean = {mean:.2f}")
    print(f"{confidence*100:.0f}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    
    # Step 4: Visualization
    plt.figure(figsize=(7,4))
    plt.errorbar(x=1, y=mean, yerr=margin, fmt='o', color='blue', capsize=10)
    plt.axhline(y=lower, color='red', linestyle='--', label='Lower Bound')
    plt.axhline(y=upper, color='green', linestyle='--', label='Upper Bound')
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.ylabel("Mean Value")
    plt.legend()
    plt.show()

# Example Data
data = np.random.normal(loc=50, scale=5, size=40)

# Call the function
confidence_interval(data)
```

## Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

```python
import numpy as np
from scipy.stats import chi2_contingency
import pandas as pd

# Create a contingency table (e.g., Gender vs. Product Preference)
data = pd.DataFrame({
    "Gender": ["Male", "Male", "Male", "Female", "Female", "Female"],
    "Preference": ["A", "B", "C", "A", "A", "B"]
})

# Create frequency table
contingency_table = pd.crosstab(data["Gender"], data["Preference"])
print("Contingency Table:\n", contingency_table, "\n")

# Perform Chi-square test of independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Display results
print(f"Chi-square Statistic = {chi2:.3f}")
print(f"Degrees of Freedom = {dof}")
print(f"P-value = {p:.4f}")
print("\nExpected Frequencies:\n", expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nReject H₀: There is a significant relationship between Gender and Preference.")
else:
    print("\nFail to Reject H₀: No significant relationship between Gender and Preference.")
```
